```python
import numpy as np
import scipy
import matplotlib as mat
from matplotlib import pyplot as plt
import sklearn
import pandas as pd
```

```python
from sklearn import model_selection

from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
dataset = "wine_dataset.txt"
nameFeatures =['class','Alcohol','Malic acid','Ash','Alcalinity of ash','Magnesium','Total phenols','Flavanoids','Nonflavanoid phenols','Proanthocyanins','Color intensity','Hue','OD280/OD315 of diluted wines','Proline']
df = pd.read_csv(dataset,names=nameFeatures) # returns a dataframe
df.head()
```

```python
df.describe()
```

```python
print(df.shape)
```

```python
df.groupby('class').size()
```

```python
data=df.values
x=data[:,1:]##contains all info other than class label
y=data[:,0]##class label 
print(x)
print(y)
```

```python
Val_size=0.30 #test size is how much rest is training
random_seed=3 #randomly chosing
X_train,X_test,Y_train,Y_test=model_selection.train_test_split(x,y,test_size = Val_size,random_state=random_seed)
deciTree = DecisionTreeClassifier()
deciTree.fit(X_train, Y_train)
predictions = deciTree.predict(X_test)
print("Accuracy on the Test Data ")
print(accuracy_score(Y_test, predictions))
print()
print()
print("Clasification Report on the Test Data")
print(classification_report(Y_test, predictions))
```
